```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score

le = preprocessing.LabelEncoder()
```

```python
y = le.fit_transform(data.Attrition_Flag)
```

```python
encode_gender = le.fit_transform(data.Gender)
encode_education = le.fit_transform(data.Education_Level)
encode_marital = le.fit_transform(data.Marital_Status)
econde_income = le.fit_transform(data.Income_Category)
encode_card = le.fit_transform(data.Card_Category)
```

```python
X = data.drop(['CLIENTNUM','Gender','Education_Level','Marital_Status',
               'Income_Category','Card_Category','Attrition_Flag'],1)
```

```python
X['Gender'] = pd.Series(encode_gender)
X['Education_Level'] = pd.Series(encode_education)
X['Marital_Status'] = pd.Series(encode_marital)
X['Income_Category'] = pd.Series(econde_income)
X['Card_Category'] = pd.Series(encode_card)
```

```python
forest_model = RandomForestClassifier()
```

```python
forest_model.fit(X,y)
```

```python
importance = forest_model.feature_importances_
```

```python
important_features_dict = {}
for x,i in zip(X.columns,importance):
    important_features_dict[x]=i
```

```python
with plt.style.context('seaborn'):
    plt.figure(figsize=(10,5))
    plt.bar(list(important_features_dict.keys()),list(important_features_dict.values()))
    plt.xticks(rotation=90)
    plt.ylabel('Importance')
    plt.title('Random Forest Classifier Feature Importance')
    plt.show()
```

```python
{k: v for k, v in sorted(important_features_dict.items(), key=lambda item: item[1],reverse=True)}
```

```python
X_important = pd.DataFrame()
X_important['Total_Trans_Amt'] = X.Total_Trans_Amt
X_important['Total_Trans_Ct'] = X.Total_Trans_Ct
X_important['Total_Revolving_Bal'] = X.Total_Revolving_Bal
X_important['Total_Ct_Chng_Q4_Q1'] = X.Total_Ct_Chng_Q4_Q1
```

```python
X_train_important, X_test_important, y_train_important, y_test_important = train_test_split(X_important, y, test_size=0.33)
```

```python
forest_model.fit(X_train_important,y_train_important)
predictions_important = forest_model.predict(X_test_important)
```

```python
print('Features With Importance Over 10%')
print('F1 Score: %s' % f1_score(y_test_important,predictions_important))
print('Accuracy Score: %s' % accuracy_score(y_test_important,predictions_important))
print(confusion_matrix(y_test_important,predictions_important))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .33)
```

```python
forest_model.fit(X_train,y_train)
predictions = forest_model.predict(X_test)
```

```python
print('All Features')
print('F1 Score: %s' % f1_score(y_test,predictions))
print('Accuracy Score: %s' % accuracy_score(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```
